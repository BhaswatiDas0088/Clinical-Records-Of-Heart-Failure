. Introduction:

      Heart failure occurs when the heart muscle doesn’t pump blood as well as it should. When this happens, blood often backs up and fluid can build up in the lungs, causing shortness of breath. Certain heart conditions gradually leave the heart too weak or stiff to fill and pump blood properly. These conditions include narrowed arteries in the heart and high blood pressure .
      Proper treatment may improve the symptoms of heart failure and may help some people live longer. Lifestyle changes can improve quality of life. Try to lose weight, exercise, use less salt and manage stress.
      But heart failure can be life-threatening. People with heart failure may have severe symptoms. Some may need a heart transplant or a device to help the heart pump blood.
      Heart failure is sometimes called congestive heart failure.

. Purpose of Analysis:

1. Patient Care: It helps healthcare providers track the progression of the disease, monitor treatment effectiveness, and make informed decisions about patient care.

2. Research: Clinical record analysis contributes to medical research by providing insights into the factors influencing heart failure development, progression, and outcomes. This data can be used to identify patterns, risk factors, and potential interventions.

3. Quality Improvement: By examining clinical records, healthcare institutions can identify areas for improvement in heart failure management protocols, patient education, and healthcare delivery processes.

4. Resource Allocation: Understanding the patterns and trends in heart failure cases can aid in allocating resources effectively, such as personnel, equipment, and medications, to meet the needs of the patient population.

5. Outcome Prediction: Analysis of clinical records can help predict patient outcomes, such as readmission rates, mortality risk, and quality of life, enabling proactive interventions to improve outcomes.

Overall, analyzing heart failure clinical records is essential for improving patient care, advancing medical knowledge, and optimizing healthcare delivery for individuals affected by this condition.

Analysing The Dataset Through Data Visualisation

> Heart Failure Rate In Different Age Groups

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_Clinical Records of Heart Failure.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(10)
customerDf['age'].groupby(customerDf['time']).sum().plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Time')
plt.title('Heart Failure In Different Ages')
plt.show()

```

> Analysing Causes of Heart Failure By Smoking

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
filepath2 = 'after_drop_Clinical Records of Heart Failure.csv'
studentDf = pd.read_csv(filepath2)
studentDf.dropna(inplace=True)
studentDf.drop_duplicates(inplace=True)
mycolor=np.array(['yellow','lightgreen'])
studentDf['smoking'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=mycolor)
plt.title("Causes Of Heart Failure By Smoking")
plt.legend(title='Causes')
plt.show()
```

> Analysing High Blood Pressure By Their Ages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_Clinical Records of Heart Failure.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(5)
mycolor=np.array(['orange'])
customerDf['high_blood_pressure'].groupby(customerDf['age']).sum().sort_values(ascending=False).plot(kind='barh',color=mycolor)
plt.xlabel('High Blood Pressure')
plt.ylabel('Age')
plt.title(' High Blood Pressure By Their Ages')
plt.show()
```

> Analysing Diabetes causes Of Heart Failure

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath = 'after_drop_Clinical Records of Heart Failure.csv'
customerDf=pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure(figsize=(6,5))
mycolor=np.array(['red','skyblue'])
customerDf['diabetes'].value_counts().plot(kind='pie',autopct='%1.0f%%',colors=mycolor)
plt.title('Rate Of Diabetes')
plt.legend(title='Diabetes')
plt.show()
```

> Death Event By Ages

```python
dataset=pd.read_csv('after_drop_Clinical Records of Heart Failure.csv')
colors = [ 'pink','cyan','red','orange','green','blue','yellow']
dataset.groupby('age')['DEATH_EVENT'].sum().sort_values(ascending=False).head(10).plot(kind='barh',xlabel='Death Event',ylabel='Age',title='Death Event By Ages',color=colors)
```
